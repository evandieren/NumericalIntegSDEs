## Brownian Motion generation

```julia
# For loop version
function std_BM_loop(n::Int64,T::Float64)
    dt = T/n
    W = Array{Float64}(undef, n+1)
    d = Normal()
    dW = sqrt(dt)*rand(d,n)
    W[1] = 0.0
    for i in 2:n+1
        W[i] = W[i-1] + dW[i-1]
    end
    return W
end

# cumsum version
function std_BM_cum(n::Int64,T::Float64)
    dt = T/n
    d = Normal()
    dW = sqrt(dt)*rand(d,n)
    return [0;cumsum(dW)]
end

n = 300
T = 5.0
dt = T/n
W1 = std_BM_cum(n,T)
W2 = std_BM_loop(n,T)
x_plot = 0:dt:T
plot(x_plot,[W1,W2])
```
