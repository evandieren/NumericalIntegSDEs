## Brownian Motion generation

Consider a scalar standard Brownian motion (Wiener process) on [0, 1]

```julia
Random.seed!(1234)
```

```julia
# For loop version
function std_BM_loop(n::Int64,T::Float64)
    dt = T/n
    W = Array{Float64}(undef, n+1)
    dW = sqrt(dt)*rand(Normal(),n)
    W[1] = 0.0
    for i in 2:n+1
        W[i] = W[i-1] + dW[i-1]
    end
    return W
end

# cumsum version
function std_BM_cum(n::Int64,T::Float64)
    dt = T/n
    dW = sqrt(dt)*rand(Normal(),n)
    W = [0;cumsum(dW)]
    return W
end

n = 300
T = 5.0
dt = T/n
@btime W1 = std_BM_cum(n,T)
@btime W2 = std_BM_loop(n,T)

x_plot = 0:dt:T
plot(x_plot,[W1,W2])
```
